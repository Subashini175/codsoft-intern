```python
import plotly.graph_objects as go
import pandas as pd
df = pd.DataFrame(data)
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['Strike Rate']))
fig.update_layout(title='Team Performance Over Seasons',
                 xaxis_title='Batter id',
                 yaxis_title='Strike rate',
                 font=dict(size=14))
fig.show()
```
